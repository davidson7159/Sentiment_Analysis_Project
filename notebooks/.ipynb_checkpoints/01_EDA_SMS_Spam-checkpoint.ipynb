### Import - librairies and dataset

```python
# 01_EDA_SMS_Spam.ipynb
# SMS Spam Classification - Exploratory Data Analysis

# -------------------------------
# 1️⃣ Import libraries
# -------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
import string

import warnings
warnings.filterwarnings('ignore')
# download stopwords for NLTK
nltk.download('stopwords')

# -------------------------------
# 2️⃣ Load dataset
# -------------------------------
# Replace with your actual file name
df = pd.read_csv("/Users/adriendavidson/Documents/Davidson/sms-spam-nlp/spam.csv", usecols = [0, 1], encoding = 'latin-1')  

# Quick look
df.head()
```

### Dataset overview

```python
# 3️⃣ Basic info
# -------------------------------
print("Dataset shape:", df.shape)
print("\nColumns:", df.columns)
print("\nMissing values:\n", df.isnull().sum())
```

### Target distribution

```python
# 4️⃣ Check spam vs ham distribution
# -------------------------------

# Proportions
print(df['v1'].value_counts(normalize=True))


ax = sns.countplot(data=df, x='v1')
for container in ax.containers:
    ax.bar_label(container, padding = 1)
plt.title("Spam vs Ham distribution")
plt.show()
```

### Message length analysis

```python
# 5️⃣ Message length analysis
# -------------------------------
plt.figure(figsize=(8, 6))

df['message_length'] = df['v2'].apply(len) 
sns.histplot(df['message_length'], bins=50)
plt.title("Message Length Distribution")
plt.show()

# spam vs ham lengths
plt.figure(figsize=(8, 6))

sns.boxplot(x='v1', y='message_length', data=df)
plt.title("Message Length by Label")
plt.show()
```

### Most common words: ham vs spam

```python
# 6️⃣ Most common words
# -------------------------------
from collections import Counter

# Preprocessing function
def preprocess_text(text):
    text = text.lower()
    text = ''.join([c for c in text if c not in string.punctuation])
    return text

df['clean_text'] = df['v2'].apply(preprocess_text)

# Split spam and ham
spam_words = ' '.join(df[df['v1']=='spam']['clean_text']).split()
ham_words = ' '.join(df[df['v1']=='ham']['clean_text']).split()

# Remove stopwords
stop_words = set(stopwords.words('english'))
spam_words = [w for w in spam_words if w not in stop_words]
ham_words = [w for w in ham_words if w not in stop_words]

# Count most common words
spam_counts = Counter(spam_words).most_common(20)
ham_counts = Counter(ham_words).most_common(20)

print("Top 20 spam words:", spam_counts)
print("Top 20 ham words:", ham_counts)
```
